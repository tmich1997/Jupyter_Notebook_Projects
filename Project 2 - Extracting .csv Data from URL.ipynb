# Project 2 - Extracting CSV Data from URL

```python
# reading a .csv file from a website (url)
# target website https://www.football-data.co.uk/englandm.php
import pandas as pd
df_premier23 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2324/E0.csv')
```

```python
df_premier23
```

```python
# renaming field names
# use dictionary 
# rename(columns = {'original name of field #1': 'new name of field #1', 'original name of field #2': 'new name of field #2'}, inplace = True) 
df_premier23.rename(columns = {'FTHG':'home_goals', 
                               'FTAG':'away_goals'}, inplace = True)
```
